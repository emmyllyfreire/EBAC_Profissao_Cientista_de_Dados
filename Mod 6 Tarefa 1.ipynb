#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sinasc = pd.read_csv('SINASC_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(5,1, figsize=(8,16)) 

ticks = [0, 1, 2, 3, 4, 5]
labels_gestacao = ['37 a 41', '>42', '32 a 36', '22 a 27', '28 a 31','<22']

sns.countplot(x='APGAR1', data=sinasc, ax=axes[0])
sns.countplot(x='APGAR5', data=sinasc, ax=axes[1])
sns.countplot(x='SEXO', data=sinasc, ax=axes[2])
sns.countplot(x='GESTACAO', data=sinasc, ax=axes[3])
sns.countplot(x='ESTCIVMAE', data=sinasc, ax=axes[4])

axes[3].tick_params(axis='x', labelsize=10)
axes[3].set_xticks(ticks, labels=labels_gestacao)
axes[3].set_xlabel('GESTACAO (semanas)')

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(8, 6)) 
sns.barplot(y='APGAR5', x='CONSULTAS', data=sinasc)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2,2) 

sns.countplot(x='SEXO', data=sinasc, ax=axes[0,0])
sns.countplot(x='GESTACAO', data=sinasc, ax=axes[0,1])
sns.countplot(x='ESTCIVMAE', data=sinasc, ax=axes[1,0])
sns.countplot(x='CONSULTAS', data=sinasc, ax=axes[1,1])

axes[0,1].tick_params(axis='x', labelsize=8,  rotation=90)
axes[0,1].set_xticks(ticks, labels=labels_gestacao)
axes[0,1].set_xlabel('GESTACAO (semanas)')


plt.subplots_adjust(wspace=0.9, hspace=0.9)
```
